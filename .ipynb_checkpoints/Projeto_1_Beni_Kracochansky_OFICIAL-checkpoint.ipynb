## Análise de dados

Ao dividir as variáveis nas regiões já previamente estabelecidas, isso torna mais fácil a análise, já que foi sinalizada uma variável a cada região e assim um dataframe para cada uma. Após isso, foi feita uma tabela dinâmica a qual se analisa o desvio padrão das variáveis. Vale ressaltar que este dado é extremamente importante para o estudo das três variáveis divididas nas regiões escolhidas. O desvio padrão é uma medida que expressa o grau de dispersão de um conjunto de dados. Ou seja, o desvio padrão indica o quanto um conjunto de dados é uniforme. Quanto mais próximo de 0 for o desvio padrão, mais homogêneo são os dados. Mas por que é tão importante trabalhar com o desvio padrão neste caso em específico?

Neste caso, estamos trabalhando com uma quantidade muito grande de países e categorizamo-os em grupos distintos. Os países que participam da União Européia tendem a ter um desvio padrão menor na maioria dos dados (exceto em PIB, o que será discutido mais adiante) por serem parte de um bloco econômico e por ser em menor quantidade, assim as chances de terem países com extrapolações, ou que 'saem muito da curva' diminui. Por outro lado, quando formos analisar os países que estão forma da Europa, há uma tendência destes possuírem um desvio padrão maior, por estarem comparando mais países, e não só isso mas também países de diferente calibre econômico e político. Por exemplo, na mesma categoria estamos dizendo de comparar uma das maiores potências econômicas mundiais (Estados Unidos) e países mais pobres do mundo situados na África e Ásia. 

Por isso, foi criada uma terceira categoria que serve para analisar os países que estão na Europa mas ainda assim não participam da União Européia. 

```python
# Dividindo as variáveis e sinalizando-as nas divisões de regiões já estabelecidas.

un = dataframe[dataframe.EU == 'Participa']
eur = dataframe[dataframe.EU == 'Não participa']
ne = dataframe[dataframe.EU == 'Não europeia']

pd.pivot_table(data=dataframe, index='EU', aggfunc=np.std)
```

Em seguida foram definidas médias para cada ano de cada variável. Esse processo foi feito para poder começar a trabalhar com quantidades de dados menores, ou seja, comecei a resumir os dados em um número só ao invés de vários diferentes. Por causa da sinalização de cada variável a uma região, foi possível refinar ainda mais a análise, onde foi criado uma variável para cada média de cada variável em cada região. Tudo isso em uma só variável.

Para driblar o problema de considerar médias como os próprios dados, mais adiante do trabalho foram trabalhadas com a mediana e a média juntas em um só gráfico, pelo qual seria possível analisar se é válido de fato analisar a média ou se a mediana. Mais adiante ainda, foi trabalhada as correlações entre as variáveis quantitativas, o que possibilitou a análise ainda mais pertinente à pergunta.

```python
## Definindo as médias de cada ano para cada variável quantitativa dividida nos três dataframes.

mean_inv_un_2010 = un['2010_Investments'].mean()
mean_inv_eur_2010 = eur['2010_Investments'].mean()
mean_inv_ne_2010 = ne['2010_Investments'].mean()

mean_inv_un_2011 = un['2011_Investments'].mean()
mean_inv_eur_2011 = eur['2011_Investments'].mean()
mean_inv_ne_2011 = ne['2011_Investments'].mean()

mean_inv_un_2012 = un['2012_Investments'].mean()
mean_inv_eur_2012 = eur['2012_Investments'].mean()
mean_inv_ne_2012 = ne['2012_Investments'].mean()


mean_gdp_un_2010 = un['2010_GDP'].mean()
mean_gdp_eur_2010 = eur['2010_GDP'].mean()
mean_gdp_ne_2010 = ne['2010_GDP'].mean()

mean_gdp_un_2011 = un['2011_GDP'].mean()
mean_gdp_eur_2011 = eur['2011_GDP'].mean()
mean_gdp_ne_2011 = ne['2011_GDP'].mean()

mean_gdp_un_2012 = un['2012_GDP'].mean()
mean_gdp_eur_2012 = eur['2012_GDP'].mean()
mean_gdp_ne_2012 = ne['2012_GDP'].mean()


mean_trab_un_2010 = un['2010_Trab'].mean()
mean_trab_eur_2010 = eur['2010_Trab'].mean()
mean_trab_ne_2010 = ne['2010_Trab'].mean()

mean_trab_un_2011 = un['2011_Trab'].mean()
mean_trab_eur_2011 = eur['2011_Trab'].mean()
mean_trab_ne_2011 = ne['2011_Trab'].mean()

mean_trab_un_2012 = un['2012_Trab'].mean()
mean_trab_eur_2012 = eur['2012_Trab'].mean()
mean_trab_ne_2012 = ne['2012_Trab'].mean()

```

# Análise do PIB 

Todo estudo foi feito com base na comparação com o PIB médio das 3 regiões pré estabelecidas através dos 3 anos analisados. Assim, para evitar equívocos, foi plotado um gráfico do PIB pelos anos que passam, e deste modo, conseguimos visualizar a situação em um panorama diferente. Analisando, é percebido que a média do PIB dos países da União Européia tende a ser maior que a média dos outros. Em seguida vem os países não europeus e depois os europeus não participantes do bloco econômico. Este gráfico nos permite não só ter um panorama de qual a situação econômica média de cada região mas também ver que as três estão em um crescimento, não necessariamente contínuo.

```python
anos = [2010,2011,2012]
total_mean_gdp_un = [mean_gdp_un_2010, mean_gdp_un_2011, mean_gdp_un_2012]
total_mean_gdp_eur = [mean_gdp_eur_2010, mean_gdp_eur_2011, mean_gdp_eur_2012]
total_mean_gdp_ne = [mean_gdp_ne_2010, mean_gdp_ne_2011, mean_gdp_ne_2012]

plt.scatter (anos,total_mean_gdp_un, label = 'União Europeia', color = 'yellow')
z = np.polyfit(anos, total_mean_gdp_un, 1)
p = np.poly1d (z)
plt.plot (anos,p(anos),'y--')

plt.scatter (anos,total_mean_gdp_ne, label = 'Não Europeus', color = 'blue')
z = np.polyfit(anos, total_mean_gdp_ne, 1)
p = np.poly1d (z)
plt.plot (anos,p(anos),'b--')

plt.scatter (anos,total_mean_gdp_eur, label = 'Europeus não participantes', color = 'green')
z = np.polyfit(anos, total_mean_gdp_eur, 1)
p = np.poly1d (z)
plt.plot (anos,p(anos),'g--')

plt.title ('PIB através dos 3 anos')
plt.legend (loc = 'center left', bbox_to_anchor = (1,0.5))
plt.ylabel (' PIB (ajustado à inflação) em USD')
plt.grid (True)
plt.xlabel ('Anos')
plt.show()
```

```python
# Fazendo médias das três variáveis através dos 3 anos estudados.

mean_inv_un = (mean_inv_un_2010 + mean_inv_un_2011 + mean_inv_un_2012)/3 
mean_inv_eur = (mean_inv_eur_2010 + mean_inv_eur_2011 + mean_inv_eur_2012)/3
mean_inv_ne = (mean_inv_ne_2010 + mean_inv_ne_2011 + mean_inv_ne_2012)/3

mean_gdp_un = (mean_gdp_un_2010 + mean_gdp_un_2011 + mean_gdp_un_2012)/3
mean_gdp_eur = (mean_gdp_eur_2010 + mean_gdp_eur_2011 + mean_gdp_eur_2012)/3
mean_gdp_ne = (mean_gdp_ne_2010 + mean_gdp_ne_2011 + mean_gdp_ne_2012)/3

mean_trab_un = (mean_trab_un_2010 + mean_trab_un_2011 + mean_trab_un_2012)/3
mean_trab_eur = (mean_trab_eur_2010 + mean_trab_eur_2011 + mean_trab_eur_2012)/3
mean_trab_ne = (mean_trab_ne_2010 + mean_trab_ne_2011 + mean_trab_ne_2012)/3
```

Abaixo está plotado um gráfico de barras que representa a média do PIB de cada região, ou seja, o mesmo dado que está no gráfico de cima mas em uma forma de gráfico diferente.

```python
mean_gdp = {'União Européia': mean_gdp_un , 'Europeus não participantes': mean_gdp_eur , 'Não Europeus': mean_gdp_ne}
names=list(mean_gdp.keys())
values=list(mean_gdp.values())

plt.bar(names,values, color = ('blue','turquoise','yellow'))
plt.title("PIB")
plt.ylabel('PIB total em USD (ajustado à inflação)')
```

A análise mais específica começa a tomar forma a partir daqui. Abaixo estão plotados alguns histogramas, estes com a frequência absoluta em seu eixo y e o PIB em seu eixo x. O histograma é um tipo de gráfico de frequência que tem por objetivo ilustrar como uma determinada amostra ou população de dados está distribuída. É normal visualizar que os números dos eixos y variam de acordo com a mudança de cor de gráfico pois cada cor representa uma região, e assim cada região possui um número de países diferentes e consequentemente, uma frequência absoluta diferente.
Mesmo que os gráficos abaixo possuem a mesma faixa e intervalo no eixo x, isso não é o ideal para a análise dos histogramas dos países que não estão na Europa. Por isso, mais abaixo estão plotados os gráficos dessa região sob uma faixa onde é mais fácil de analisar suas componentes e características.

Existem três formas de caracterizar um histograma quanto à distribuição de dados no seu intervalo. Estes são a simetria, a assimetria à direita e a assimetria à esquerda. A simetria normal ocorre quando a média e a mediana possuem o mesmo valor, e assim é posssível afirmar que um conjunto de dados possui exatamente a mesma quantidade distribuída depois e antes de sua média e mediana, fazendo assim, uma forma simétrica. A assimetria à esquerda acontece quando a média possui um valor menor que a mediana, fazendo assim o gráfico ter uma  cauda maior para a esquerda. Por último, a assimetria à direita é quando a média fica à frente da mediana. Curiosamente, **todos os gráficos plotados demonstram ter uma assimetria à  direita**. Mas o que isso significa? Isso significa que certos países, mesmo que em minoria, possuem um PIB tão alto que conseguem puxar a média para mais prósimo deles e mais distante da mediana. É evidente que a distância entre a média e a mediana é maior no caso de países que não estão na Europa. Isso acontece por dois motivos. O primeiro deles é porque estamos tratando de uma série de dados com um 'n' maior, ou seja, mais países estão sendo analisados. O segundo motivo é porque diferentes países estão sendo analisados e estes se situam em situações econômicas muito divergentes. Resultado disso é que países como Estados Unidos, com um PIB bastante elevado, carregam os outros e acaba distanciando a média da mediana.

Comparando países europeus não participantes da União Europeia com países participantes do bloco leva-se a duas conclusões. A primeira é evidente, que eles tratam de valores menores, ou seja PIBs menores e sob uma contagem de países menor que a própria UE. A segunda é que a distância de sua média para a mediana é evidentemente menor que a da União Européia, levantando certas questões para se entrar na União Europeia realmente é sinônimo de sucesso. Mesmo que essa comparação não faça sentido de imediato, é possível analisar que quando pensamos em países da UE diretamente pensamos que eles fazem parte da 'elite' e fazem mais dinheiro do que aqueles que não participam do bloco. Mesmo que isso possa ser verdade, quando olhamos os gráficos abaixo, essa informação pode ser desconsiderada, já que do mesmo jeito que tem países que possuem mais PIB e estão fora da UE, também tem países dentro da UE com um PIB menor que o de quem está fora.

```python
def verticalLine(plot, x, y, color, nome):
    plot.axvline(x, color=color, linestyle="dashed", linewidth=2)
    plot.text(1.01*x, y, nome, color=color) 

fig = plt.figure(figsize=(15, 15))
plt.subplot(331)
plot = un['2010_GDP'].plot.hist(bins=5, title='União Européia: PIB 2010', alpha=0.9)
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, un['2010_GDP'].mean(), 15, "y", "Mean")
verticalLine(plot, un['2010_GDP'].median(), 12, "g", "Median")

plt.subplot(332)
plot = un['2011_GDP'].plot.hist(bins=5, title='União Européia: PIB 2011', alpha=0.9)
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, un['2011_GDP'].mean(), 15, "y", "Mean")
verticalLine(plot, un['2011_GDP'].median(), 12, "g", "Median")

plt.subplot(333)
plot = un['2012_GDP'].plot.hist(bins=5, title='União Européia: PIB 2012', alpha=0.9)
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, un['2012_GDP'].mean(), 15, "y", "Mean")
verticalLine(plot, un['2012_GDP'].median(), 12, "g", "Median")

plt.subplot(334)
plot = eur['2010_GDP'].plot.hist(bins=5, title='Europeus não participantes: PIB 2010', alpha=0.9, color = 'turquoise')
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, eur['2010_GDP'].mean(), 5, "y", "Mean")
verticalLine(plot, eur['2010_GDP'].median(), 3, "g", "Median")

plt.subplot(335)
plot = eur['2011_GDP'].plot.hist(bins=5, title='Europeus não participantes: PIB 2011', alpha=0.9, color = 'turquoise')
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, eur['2011_GDP'].mean(), 5, "y", "Mean")
verticalLine(plot, eur['2011_GDP'].median(), 3, "g", "Median")

plt.subplot(336)
plot = eur['2012_GDP'].plot.hist(bins=5, title='Europeus não participantes: PIB 2012', alpha=0.9, color = 'turquoise')
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, eur['2012_GDP'].mean(), 5, "y", "Mean")
verticalLine(plot, eur['2012_GDP'].median(), 3, "g", "Median")

plt.subplot(337)
plot = ne['2010_GDP'].plot.hist(bins=5, title=' Não Europeus: PIB 2010', alpha=0.9, color='salmon')
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, ne['2010_GDP'].mean(), 50, "y", "Mean")
verticalLine(plot, ne['2010_GDP'].median(), 40, "g", "Median")

plt.subplot(338)
plot = ne['2011_GDP'].plot.hist(bins=5, title=' Não Europeus: PIB 2011', alpha=0.9, color='salmon')
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, ne['2011_GDP'].mean(), 50, "y", "Mean")
verticalLine(plot, ne['2011_GDP'].median(), 40, "g", "Median")

plt.subplot(339)
plot = ne['2012_GDP'].plot.hist(bins=5, title=' Não Europeus: PIB 2012', alpha=0.9, color='salmon')
plt.ylabel('Frequência Absoluta')
plt.xlabel('PIB em USD (ajustado à inflação)')
plot.set_xlim([0,4*1e12])
verticalLine(plot, ne['2012_GDP'].mean(), 50, "y", "Mean")
verticalLine(plot, ne['2012_GDP'].median(), 40, "g", "Median")

plt.show()
```

Em seguida, a análise que busca propriamente responder a minha pergunta, de qual das duas variáveis (horas de trabalho, ou porcentagem que o investimento representa do PIB) pesa mais no PIB. Decidi calcular isso através do método de correlação e ainda plotar gráficos que apresentam os pontos de dispersão e um linha de tendência não de primeiro grau, através do comando '.regplot'. Esse comando também me permitiu analisar de forma mais eficiente meus dados, já que ele preenche com uma sombra o espaço onde há uma chance de 95% de meus dados estarem presentes. Isso significa que se eu possuir um conjunto de dados com pontos mais dispersos, a chance da sombra ser maior aumenta, enquanto caso for o contrário, a sombra diminui.

A análise ocorreu da seguinte forma. Primeiramente comparou-se todas as variáveis em todos os anos com os PIBs correspondentes, em cada região estipulada. Também é possível visualizar o número resultante da correlação. Lembrando que ela varia de -1 a 1, ou seja, quanto mais perto destes extremos, mais correlacionada é a variável, ou seja, mais 'dependente' uma é da outra. Ao ver os gráficos é possível de fato ver que nenhuma possui uma correlação forte, e o máximo que observamos é uma série de correlação fraca negativa (todos os anos na comparação da União Européia entre a quantidade de horas trabalhadas com o PIB).

# Conclusão

Com base nos dados analisados, a conclusão é de que mesmo que ambas correlações não apresentem de fato 'peso' ou quaisquer influência no PIB total, como analisado em uma parte do código, a única variável que conseguiu atingir qualquer tipo de correlação foi a quantidade de horas trabalhadas semanalmente em países da União Européia. Mesmo que a maior parte da análise matemática leva em consideração a comparação das variáveis separadamente, ainda é pertinente compará-las e analisa-las de forma conjunta. Ainda que matematicamente essa seja a conclusão que foi chegada, se analisarmos de forma mais geral, nenhuma das duas variáveis possui um peso tão considerável no PIB. Por que?


Uma possível explicação a este fato pode ser porque não foram analisadas as variáveis corretas. Talvez o investimento realmente seja uma variável quantitativa que não é muito pesada em nenhum país, e como já foi visto, o PIB é composto por muitos fatores, e o investimento pode ser que nunca chegue a ser um dos que mais influencia o aumento ou não do índice. Resumindo, independentemente de quão grande for o fator de investimento no PIB, pode ser que sempre terão outros que serão prioridades antes de chegar neste índice, ou seja, o investimento acaba sendo uma consequência da movimentação da economia no país, o que nos leva a pensar que semore vai ter outro fator sendo mais importante do que o investimento.


Explicando a não correlação da segunda variável, novamente, como a primeira, pode ser que a variável não seja a melhor para analisar o que a pergunta precisa como resposta. A quantidade de horas que um indivíduo trabalha semanalmente, ao analisar, pode variar muito de um lugar do mundo para outro. Alguém que trabalha 30 horas em um país não participante da União Europeia não necessariamente contribui mais para o PIB do que alguém que trabalha 60 e que está na União Européia. Assim, de fato a cultura do país e a relação da economia com o mercado de trabalho são variáveis muito importantes e que não foram contabilizadas e levadas em conta neste estudo.


Respondendo a pergunta de uma vez por todas, se comparando as duas variáveis quantitativas (Horas trabalhadas ou Investimento) há, em algum grau um tipo de influência quando correlacionada ao PIB, e se o país estar na União Europeia ou não afeta este dado, a resposta é não (parcialmente). Mesmo que muito baixa, pudemos ver que há uma correlação quando analisamos os países da União Européia e a média de horas trabalhadas em uma semana. Respondendo a segunda parte da pergunta, se o país estar ou não no bloco econômico afeta essa resposta, sim, afeta. Vimos que há a existência da correlação, mesmo que negativa quando se analisa o bloco econômico europeu, mas quando se vai comparar àqueles que não participam, o resultado muda, levando-nos a concluir que sim, fazer ou não parte da UE faz diferença quando analisamos a influência  destas variáveis no PIB.
